# Activation Patching

Let's us modify activations at specific model components to see how they causally affect the model's output.

## Sequence

1. Baseline Run.
    - Run the model and record its activations
2. Corrupted Run.
    - Run the model with a counterfactual (i.e., corrputed) prompt and record the difference in activations.
3. Patching
    - Replace activations at the model component of interest with alternate activations (or zeros, which is sometimes referred to as ablation).

## Why

Doing all this lets us get an idea of the information flow in the model.

A common reason to do this is to identify circuits, subgraphs in the model that are responsible for some interpretable behavior.
